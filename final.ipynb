```python
# converting object to int
con_obj=['humidity','wind_speed','pressure','string_id','error_code','installation_type']
for col in con_obj:
    dataset[col] = pd.to_numeric(dataset[col],errors='coerce')
```

```python
# converting object to int
con_obj2=['humidity','wind_speed','pressure','string_id','error_code','installation_type']
for col in con_obj2:
    dataset2[col] = pd.to_numeric(dataset2[col],errors='coerce')
```

```python
#replacing null values with mean or median

mean_median=['temperature','irradiance','panel_age','maintenance_count','soiling_ratio',
             'voltage','current','module_temperature','cloud_coverage']
for col in mean_median:
    skewness=dataset[col].skew()
    if abs(skewness)>0.5 :
        dataset[col]=dataset[col].fillna(dataset[col].median())
    else:
        dataset[col]=dataset[col].fillna(dataset[col].mean())
```

```python
skewness=dataset['temperature'].skew()
if abs(skewness)>0.5 :
    dataset[col]=dataset[col].fillna(dataset[col].median())
else:
    dataset[col]=dataset[col].fillna(dataset[col].mean())
```

```python
# dropping id column
dataset.drop(columns=['id'], inplace=True)
dataset2.drop(columns=['id'], inplace=True)
```

```python
# Separate target variable from training data
x_train = dataset2.drop(columns=['efficiency'])
y_train = dataset2['efficiency']
```

```python
# Identify numerical and categorical columns
numerical_cols = x_train.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols = x_train.select_dtypes(include=['object']).columns.tolist()
```

```python
x_test=dataset.copy()
```

```python
numerical_cols
```

```python
from scipy.stats import skew
```

```python
skew_vals = x_train[numerical_cols].skew().sort_values()
```

```python
skew_vals
```

```python

# Threshold to determine high skewness
high_skew = skew_vals[abs(skew_vals) > 0.5].index.tolist()
low_skew = skew_vals[abs(skew_vals) <= 0.5].index.tolist()

# Now create two pipelines accordingly
median_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])

mean_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean'))
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('median', median_pipeline, high_skew),
        ('mean', mean_pipeline, low_skew),
        ('cat', categorical_pipeline, categorical_cols)
    ]
)
```

```python
x_train_clean = preprocessor.fit_transform(x_train)
x_test_clean = preprocessor.transform(x_test)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_tr, y_val = train_test_split(x_train_clean, y_train, test_size=0.2, random_state=42)
```

```python
model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.05,
    depth=6,
    loss_function='RMSE',
    eval_metric='RMSE',
    random_seed=42,
    early_stopping_rounds=50,
    verbose=100
)
```

```python
model.fit(X_train, y_tr, eval_set=(X_val, y_val))

# Predict
y_pred = model.predict(x_test_clean)
```

```python
y_pred = y_pred.flatten()


df = pd.DataFrame({
    'id': range(len(y_pred)),
    'efficiency': y_pred
})
```
